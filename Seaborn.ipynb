**heatmap**

```python
# use for number to number value
import matplotlib.pyplot as plt
import seaborn as sns
data=sns.load_dataset("tips")
print(data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
tips = sns.load_dataset("tips")

# Create a pivot table: average total_bill per day and sex
heatmap_data = tips.pivot_table(index="day", columns="sex", values="total_bill", aggfunc="mean")

# Plot heatmap
sns.heatmap(data=heatmap_data, cmap="YlGnBu", annot=True)
plt.title("Average Total Bill by Day and Sex")
plt.show()
```

**multiple plots**

```python
import matplotlib.pyplot as plt
import seaborn as sns
data=sns.load_dataset("tips")
print(data)
```

```python
a=sns.FacetGrid(data,col="day",height=2)
a.map(sns.scatterplot,"total_bill","tip")
plt.show()
```

```python
a=sns.FacetGrid(data,col="time",height=2,hue="sex")
a.map(sns.barplot,"day","tip")
plt.show()
```
