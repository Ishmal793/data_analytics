```python
import pandas as pd
df=pd.read_excel("expense3.xlsx")
print(df)
print(df.isnull().sum())
df['Payment Mode']=df['Payment Mode'].fillna(df['Payment Mode'].mode()[0])
print(df.isnull().sum())
```

```python
group=df.groupby("Category").agg({"Amount":"sum"})
print(group)
```

```python
grouped_by = df.groupby("Category")["Amount"].sum()

plt.figure(figsize=(10,6))
plt.step(range(len(grouped_by)), grouped_by.values, where="mid", marker="o")
plt.xticks(ticks=range(len(grouped_by)), labels=grouped_by.index, rotation=45)
plt.xlabel('Category')
plt.ylabel('Total Amount')
plt.title('Step Chart: Total Amount by Category')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[45,67,33,67,12,]
y1=[41,60,13,66,13]
y2=[45,67,42,63,12,]
plt.figure(figsize=[5,3])
plt.plot(x,y,label="line1")
plt.plot(x,y1,label="line2")
plt.plot(x,y2,label="line3")
plt.legend(["male","female","other"],bbox_to_anchor=(0.8,1.1),ncols=2 ,labelspacing=2)
plt.show()
# ncols no of colums in legend
# 10 for center


plt.show()
```

**how to save it**

```python
import matplotlib.pyplot as plt
x=["part1","part2","part3","part4",'part5']
y=[98,67,88,95,88]
plt.xlabel("parts of herry potter")
plt.ylabel("populatity",fontsize=18)
color=["red","green","yellow","blue","orange"]
plt.title("Populatity of herry potter",fontsize=22)
plt.bar(x,y,color=color,edgecolor="black"   )
plt.savefig("plot.png")
plt.show()
```

```python
# pie chart
import matplotlib.pyplot as plt
brands=["oneplus","vivo","apple","nokia","redmi"]
x=[27,35,50,22,33]
c=["red","green","pink","blue","orange"]
ex=[0,0,0,0,1]
plt.pie(x,labels=brands,autopct="%1.2f%%",colors=c,explode=ex,shadow=True,startangle=180)
plt.savefig("pie.png",facecolor="black")
plt.show()
```
